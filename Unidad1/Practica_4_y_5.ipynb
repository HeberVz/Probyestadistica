# Practica 4: Tablas de frecuencias

```python
# creamos una lista vacia que representa la edad de 200 personas
import random
edades = []
# hacemos un ciclo para generar 100 numeros aleatorios que representan la edad de las personas
for i in range(100):
  # generamos un numero aleatorio con la funcion randit y lo guardamos en la variable
  # numero aleatorio
  numero_aleatorio = random.randint( 18, 70)
  # agreagamos el numero aleatorio a la lista de valores
  edades.append(numero_aleatorio)
print(edades)
```

```python
import pandas as pd
```

```python
# crear un DataFrame de pandas con los numeros aleatorios
df = pd.DataFrame(edades, columns=['Edad'])
df.head()
```

```python
# calcular la tabla de distribucion de frecuencias
tabla_frecuencias = df['Edad'].value_counts().reset_index()
tabla_frecuencias.columns = ['Edad', 'Frecuencia']
tabla_frecuencia = tabla_frecuencias.sort_values(by='Edad')

# imprimir la tabla de distribución de frecuencias
print("Tabla de Distribución de Frecuencias:")
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Relativa'] = tabla_frecuencias['Frecuencia'] / 100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Porcentual'] = tabla_frecuencias['Frecuencia'] / 100 * 100
tabla_frecuencias
```

```python
tabla_frecuencias['Frecu Acumulada'] = tabla_frecuencias['Frecuencia'].cumsum()
tabla_frecuencias
```

# Regla de sturges
1.Es una regla que sirve para calcular el número de clases o intervalos idóneo en los que se debe dividir el conjunto de datos
2.Establece que el número de clases es igual a uno más el logaritmo

```python
# importamos la libreria math para funciones matematicas
import math
```

```python
print(edades)
```

```python
# Crear un DataFrame con las edades
df = pd.DataFrame({'Edad' : edades})
df
```

```python
# Calcular el número de intervalos utilizando la regla de Sturges
n = len(edades)
k = int( 1 + math.log2(100))
print("Valor de n = ", n , "\nNúmero de intervalos = ", k)
```

```python
# Crear la tabla de frecuencia
tabla_frecuencia = pd.cut( df['Edad'], bins=k, include_lowest=True).value_counts().reset_index()
tabla_frecuencia.columns = ['Clase', 'Frec']
tabla_frecuencia
```

```python
# ordenar la tabla de frecuencia por el intervalo
tabla_fecuencia = tabla_frecuencia.sort_values(by='Clase')
tabla_frecuencia
```

```python
# generar columnas
tabla_frecuencia['Frec Rel'] = tabla_frecuencia['Frec'] / n
tabla_frecuencia['Frec Porc'] = tabla_frecuencia['Frec'] / n * 100
tabla_frecuencia['Frec Ac'] = tabla_frecuencia['Frec'].cumsum()
tabla_frecuencia['Frec Rel Ac'] = tabla_frecuencia['Frec Rel'].cumsum()
tabla_frecuencia['Frec Por Ac'] = tabla_frecuencia['Frec Porc'].cumsum()

# mostrar tabla de frecuencia
print(tabla_frecuencia)
```
